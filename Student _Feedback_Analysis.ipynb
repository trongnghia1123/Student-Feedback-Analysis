## Thống kê căn bản

### Thống kê tổng quan kết quả

```python
with open('file.csv', 'r') as file:
    reader = csv.reader(file)
    count_1 = 0
    count_0 = 0
    count_minus_1 = 0
    for row in reader:
        for value in row:
            if value == '1':
                count_1 += 1
            elif value == '0':
                count_0 += 1
            elif value == '-1':
                count_minus_1 += 1

print(f'Số lượng ý kiến Tích Cực: {count_1}')
print(f'Số lượng ý kiến Trung Lập: {count_0}')
print(f'Số lượng ý kiến Tiêu Cực: {count_minus_1}')
sum=count_1+count_0+count_minus_1
print(f'Tổng các ý kiến thu được:',sum)
```

### Thống kê kết quả của từng câu hỏi và lưu vào file Thongkedata.cvs

```python
list=[]
for i in range (1,11):
    b='Cau_ '+str(i)
    print('\n',b)
    count=df[b].value_counts()
    Cq0 = df[b].astype(str).str.count('0').sum()
    print('Ý kiến Trung Lập: '+ str(Cq0)+' ý kiến.')
    Cq_1 = df[b].astype(str).str.count('-1').sum()
    print('Ý kiến Tiêu Cực: '+ str(Cq_1)+' ý kiến.')   
    Cq1 = df[b].astype(str).str.count('1').sum()-Cq_1
    print('Ý kiến Tích Cực: '+ str(Cq1)+' ý kiến.\n')
    
    # tính toán
    sum= Cq0+ Cq1+ Cq_1
    Max = max(Cq0, Cq1, Cq_1)
    Min = min(Cq0, Cq1, Cq_1)
    per_cq0 = (Cq0/sum)*100
    per_cq1 = (Cq1/sum)*100
    per_cq_1 = (Cq_1/sum)*100

    # nhiều nhất
    if Max == Cq0:
        print('Câu '+str(i)+' có ý kiến Trung Lập nhiều nhất với '+ str(Max) +' ý kiến, chiếm '+str(per_cq0)+'%')
    elif Max == Cq1:
        print('Câu '+str(i)+' có ý kiến Tích Cực nhiều nhất với '+ str(Max) +' ý kiến, chiếm '+str(per_cq1)+'%')
    elif Max == Cq_1:
        print('Câu '+str(i)+' có ý kiến Tiêu Cực nhiều nhất với '+ str(Max) +' ý kiến, chiếm '+str(per_cq_1)+'%')
    
    # ít nhất
    if Min == Cq0:
        print('Câu '+str(i)+' có ý kiến Trung Lập ít nhất với '+ str(Min) +' ý kiến, chiếm '+str(per_cq0)+'%')
    elif Min == Cq1:
        print('Câu '+str(i)+' có ý kiến Tích Cực ít nhất với '+ str(Min) +' ý kiến, chiếm '+str(per_cq1)+'%')
    elif Min == Cq_1:
        print('Câu '+str(i)+' có ý kiến Tiêu Cực ít nhất với '+ str(Min) +' ý kiến, chiếm '+str(per_cq_1)+'%')

    list.append([Cq1,Cq0,Cq_1,per_cq1,per_cq0,per_cq_1])

# Lưu kết quả vào file csv

with open('Thongkedata.csv',mode='w',newline='',encoding='utf-8') as f:
    w = csv.writer(f)
    w.writerow(['Tich Cuc','Trung Lap', 'Tieu Cuc','Tich Cuc %','Trung Lap %','Tieu Cuc %'])
    for row in list:
        w.writerow(row)
```

### Kết quả thống kê của 5 câu hỏi phần 1 và phần 2

```python
# Thống kê 5 câu phần 1

print('Trung binh 5 cau hoi Phan 1:')
mean1_1 = df1['Tich Cuc %'].head(5).mean()
mean1__1 = df1['Tieu Cuc %'].head(5).mean()
mean1_0 = df1['Trung Lap %'].head(5).mean()

print('Tính tích cực :', mean1_1)
print('Tính tiêu cực:', mean1__1)
print('Tính trung lập:', mean1_0)

Max1=max(mean1_1,mean1_0,mean1__1)

print('giá trị lớn nhất câu hỏi phần 1 :', Max1)

# Thống kê 5 câu phần 2

mean2_1 = df1['Tich Cuc %'].tail(5).mean()
mean2_0 = df1['Trung Lap %'].tail(5).mean()
mean2__1 = df1['Tieu Cuc %'].tail(5).mean()

print('\nTrung binh 5 cau hoi phan 2:')
print('Tính tích cực :', mean2_1)
print('Tính tiêu cực:', mean2__1)
print('Tính trung lập:', mean2_0)

Max2 = max(mean2_1,mean2_0,mean2__1)

print('Giá trị lớn nhất câu hỏi phần 2: ',Max2)
```

KẾT QUẢ
- Vậy kết quả khảo sát cho câu hỏi phần 1 mang tính Trung lập là Ban tư vấn tuyển sinh có thể truyền tải đủ hoặc không đủ thông tin cho sinh viên
- Vậy kết quả khảo sát cho câu hỏi phần 2 mang tính Tích cực là Sinh viên quan tâm ban tư vấn đi các trường cấp 3 để trao đổi
